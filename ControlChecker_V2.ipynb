```python
def FFT(df, fftcase,fs, N):
    PSD = []
    PSDAC = []
    TIME = []
    plt.figure(figsize=(10,5))
    for i in range(1,len(fftcase)+1):
        df_serial_filter = df[(i-1)*30000:i*30000]
        timer = len(df_serial_filter) * 1 / fs
        Psd_data = plt.psd(df_serial_filter, NFFT=timer, Fs=fs, pad_to=N)
        Psd_ac_data = np.cumsum(Psd_data[0])
        PSD.append(Psd_data[0])
        PSDAC.append(Psd_ac_data)
        TIME.append(Psd_data[1])

    plt.show()

    plt.figure(figsize=(10,5))
    for i in range(0, len(fftcase)):
        plt.plot(TIME[i], PSD[i])
    plt.yscale('log')
    plt.xlim(0, 10)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('power spectral density')
    plt.title('spectral analysis')
    plt.show()

    plt.figure(figsize=(10,5))
    for i in range(0, len(fftcase)):
        plt.plot(TIME[i], PSDAC[i])
    plt.yscale('log')
    plt.xlim(0, 5)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('energy spectral')
    plt.title('spectral accumulate analysis')
    plt.show()
```

```python
def rangecheck(GearRatio,rate,maximun,minimun,result):
    out_of_range = []
    for i in range(len(result)):
        singlerate = round(result.loc[i,'Blade_1_pitch_rate']*GearRatio*30/3.14,2)
        singletorque =round(result.loc[i,'Blade_1_pitch_actuator_torque'] ,3)
        start = 0
        middle = int(np.floor(len(rate) / 2))
        end = len(rate) - 1
        while 1:
            if (singlerate==rate[middle]).all():
                start = middle
                end = middle
                break
            elif (singlerate>rate[middle]).all():
                start = middle
                middle = int((start+end)/2)
            else:
                end = middle
                middle = int((start+end)/2)
            if (end - middle )<=1 and (singlerate>rate[middle]).all()and(singlerate<=rate[end]).all():
                start = middle
                break
            elif ( middle -start)<=1 and (singlerate<rate[middle]).all()and(singlerate>=rate[start]).all():
                end = middle
                break
        if start==end:
            mintorque = minimun[start]
            maxtorque = maximun[start]
        else:
            mintorque = minimun[start]+(minimun[start]-minimun[end])/(rate[start]-rate[end])*(singlerate-rate[start])
            maxtorque = maximun[start]+(maximun[start]-maximun[end])/(rate[start]-rate[end])*(singlerate-rate[start])
        if singletorque>maxtorque or singletorque<mintorque:
#             print(singlerate,singletorque)
            if result.loc[i,'dlc'] not in out_of_range:
                out_of_range.append(result.loc[i,'dlc']) 
                return out_of_range
    return out_of_range
```

```python
def GridlossCheck(casename):
    circles = []
    failed = []
    notstop = []
    [flag,result] = lc.get_stat_data(casename, 'Generator variables', ['Generator_torque'])
    caselist = result.iloc[:,0].values.tolist()
    if casename== 'DLC24':
        for p in range(len(result)):
            if result.iloc[p, 0][3] == 'a':
                caselist.remove(result.iloc[p, 0])
    for case in caselist:
        mint = result.loc[result.dlc==case,'Generator_torque-min'].values[0]           
        flagseries,torque = lc.get_timeseries_data(casename,[case], 'Generator variables', ['Generator_torque'])
        flaggrid, gridloss = lc.get_timeseries_data(casename, [case], 'External Controller data 1',['CI_AlgGridLossShutdown'])
        if flaggrid and flagseries:
            ci = gridloss.CI_AlgGridLossShutdown.values.tolist()
            if 1 not in ci:
                notstop.append(case)
                tlist = (torque.Generator_torque==mint).tolist().index(True)
                if torque.Generator_torque[tlist-1]- torque.Generator_torque[tlist]<0.5* result.loc[result.dlc==case,'Generator_torque-max'].values[0]:
                    circles.append(case)
            else:
                tlist = (ci).index(1)
                if torque.Generator_torque[tlist+20]>= mint+5:
                    circles.append(case)
        else:
            failed.append(case)
    if len(circles)==0:
        print('%s所有工况均实现扭矩突变为0'%casename)
    else:
        print('%s中，以下工况未实现扭矩突变:'%casename)
        print(circles)
    if len(failed)>0:
        print('获取数据失败的工况有：')
        print(failed)
    if len(notstop)>0 and casename!=  'DLC51':
        print('没有触发电网掉电的工况有：')
        print(notstop)
```
